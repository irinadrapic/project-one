```python
import os
import pandas as pd
```

```python
# first lets work on 2016
file_path = os.path.join("datas", "new_mortality", "2016 mortality.csv")

df_2016 = pd.read_csv(file_path, encoding="UTF-8")

df_2016 = df_2016.rename(columns={"2016": "Number of Mortality"})

df_2016

```

Population data from https://www12.statcan.gc.ca/census-recensement/2016/dp-pd/prof/details/Page.cfm?Lang=E&Geo1=CSD&Code1=3520005&Geo2=PR&Data=Count&B1=All

Census metropolitan areas

```python
pop = [180518, 197059,134200, 1392609,  1321426, 164689, 151984, 403390, 747545, 194882, 
      161175, 523894, 494069, 144810, 4098927, 379848, 989567, 121721, 800296, 236481,
      160980, 126202, 295095, 212105, 406074, 205955, 121621, 5928040, 156042, 2463431,
      367770, 329144, 778489]
df_2016["Population"] = pop

# Ordering columns
df_2016 = df_2016[["Geography", "Age group", "Population", "Number of Mortality"]]

# calculate mortality rate for each city
mort_rate = ((df_2016["Number of Mortality"]/df_2016["Population"])*100).map("{:,.2f}%".format)
df_2016["Mortality Rate"] = mort_rate
```
