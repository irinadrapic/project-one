```python
import csv, os
import pandas as pd
import numpy as np
```

```python
house_inc_path = os.path.join("datas", "household_income", "household_inc.csv")
house_inc_row_path = os.path.join("datas","household_income", "Geo_starting_row_CSV.csv")

with open(house_inc_row_path, encoding="UTF-8") as file:
    rows = csv.reader(file)
    header = next(rows)
    
    cities = ["St. John's", "Halifax", "Moncton", "Saint John", "Fredericton",
             "Saguenay", "Québec", "Sherbrooke","Trois-Rivières",  "Drummondville",
             "Granby","Montréal", "Ottawa - Gatineau", "Kingston", "Belleville",
             "Peterborough", "Oshawa", "Toronto", "Hamilton", "St. Catharines - Niagara",
             "Kitchener - Cambridge - Waterloo", "Brantford", "Guelph", 
             "London", "Windsor", "Sarnia", "Barrie", "North Bay", "Greater Sudbury",
              "Sault Ste. Marie", "Thunder Bay", "Winnipeg","Regina", "Saskatoon",
              "Medicine Hat","Lethbridge", "Calgary","Red Deer","Edmonton","Grande Prairie",
              "Wood Buffalo","Kelowna","Kamloops","Chilliwack","Abbotsford - Mission",
              "Vancouver","Victoria","Nanaimo","Prince George",
             ]
    row_info_list = []
    for row in rows:
        
        # take only list with city name         
        if row[1] in cities:
            row_info_list.append(row)
```

```python
with open(house_inc_path, encoding='UTF-8') as file:
    inc_data = csv.reader(file)
    
    # for now, lets skip the header
    header = next(inc_data)
    
    # change all row(list) to numpy array     
    all_data = []
    for row in inc_data:
        # put all array in to a list.   
        all_data.append(np.asarray(row))
```
