```python
import pandas as pd
import os
```

```python
mort_data = os.path.join("datas","mortality", "mortality.csv")
mort_M_data = os.path.join("datas","mortality", "mortality_MetaData.csv")

mort_df = pd.read_csv(mort_data)
mort_df.head()
```

```python
# count total number of Mortaility and Potential years of life lost
count = mort_df["Indicators"].value_counts()

print(f"there are {count['Mortality']} mortality and {count['Potential years of life lost']} number of Potential years lost")
```

```python
# Drop columns that have NaN for "all" rows.
mort_df = mort_df.dropna(axis="columns", how="all")

# Just want to find number of mortality for each GEO(region) on each REF_DATE
new_mort_df = mort_df[["REF_DATE", "GEO", "Indicators"]]
new_mort_df.head()
```

```python
# count total number of Mortaility and Potential years of life lost
count = mort_df["Indicators"].value_counts()
print(count)
print()
print(f"There are {count['Mortality']} mortality and {count['Potential years of life lost']} number of Potential years lost")
```

```python
# Now that we know there are only 2 indicators mortality and Potential years lost
# replace them with a number 1

new_mort_df["Indicators"] = new_mort_df["Indicators"].replace("Mortality", 1)
new_mort_df["Indicators"] = new_mort_df["Indicators"].replace("Potential years of life lost", 1)
new_mort_df
```

```python
# see how many mortality there are in each REF_DATE
new_mort_df.groupby(["REF_DATE","GEO"]).sum()
```
