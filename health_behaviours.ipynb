```python
import pandas as pd
import os
```

```python
#Import the origina CSV
behaviours = pd.read_csv("health_behaviours.csv")
behaviours.head()
```

```python
#Organize the data by city and behaviour as well as renaming columns 

behaviours1= behaviours.drop(['REF_DATE', 'Age group','Characteristics', 'Sex'], axis=1)
behaviours1 = behaviours1.rename(columns={"GEO":"City","Indicators":"Health Behaviours","VALUE":"Number of Persons"})
behaviours1.sort_values("City")
df_new = behaviours1.set_index(["City"])
df_new["Number of Persons"] = df_new["Number of Persons"].fillna(0)
df_new["Number of Persons"] = df_new["Number of Persons"].astype(int)
df_new
```

```python
# Read the population data
population = pd.read_csv("New_population_by_city_base_clean_up.csv")
population.head()
```

```python
# Drop empty row and show only 2015 population
population_new = population.drop(population.index[0])
population_final = population_new.drop(['Age group8', '2010','2011', '2012','2013','2014','2016','2017'], axis=1)
population_final = population_final.rename(columns={'Geography':'City','2015':'Total Population'})
population_final
```

```python
# Merge the data sets on city
health_data = df_new.merge(population_final,how='outer',on='City')
#health_data = health_data.set_index(["City","Health Behaviours"])
health_data
```

```python
#Add a new column and calculate the rate of number of persons
health_data["Total Population"] = health_data["Total Population"].str.replace(",","")
health_data["Total Population"] = health_data["Total Population"].astype(int)

health_data["% Rate"] = (health_data["Number of Persons"]/health_data["Total Population"])*100
health_data["% Rate"] =  health_data["% Rate"].map("%{:.2f}".format)
health_data
```

```python
# Re-structure the data to have the cities as columns 
health_new = health_data.pivot(index='City',columns="Health Behaviours", values='% Rate')
health_new
```
